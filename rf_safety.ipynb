```python
%matplotlib inline

"""
Plotting of IEEE RF Safety Guidelines based on tables and curves in IEEE Std C95.1, 1999 Edition and 2005 Editions.
The code also overlaps 1999 and 2005 editions for both controlled and uncontrolled (public) environments.

Trigger event idea and auto-scaling of axes based on:
http://matplotlib.org/examples/subplots_axes_and_figures/fahrenheit_celsius_scales.html
"""
import math
import numpy as np
import matplotlib.pyplot as plt

def convert_E_to_S(ax1):
    """
    Scale/update power density axis according to E-field axis.
    """
    y1, y2 = ax1.get_ylim() # get y-axis range of E-field (ax1)
    
    # scale the y-axis range of the Power Density axis (ax3)
    ax3.set_ylim(math.pow(y1,2)/(120*math.pi*10), math.pow(y2,2)/(120*math.pi*10))
    ax3.figure.canvas.draw() # update plot

def convert_E_to_H(ax1):
    """
    Scale/update H-field axis according to E-field axis.
    """
    y1, y2 = ax1.get_ylim()
    
    # scale the y-axis range of the H-field axis (ax2)
    ax2.set_ylim(y1/(120*math.pi), y2/(120*math.pi))
    ax2.figure.canvas.draw() # update plot

def convert_E_to_S_U(ax4):
    """
    Scale/update power density axis according to E-field axis.
    """
    y1, y2 = ax4.get_ylim() # get y-axis range of E-field (ax1)
    
    # scale the y-axis range of the Power Density axis (ax3)
    ax6.set_ylim(math.pow(y1,2)/(120*math.pi*10), math.pow(y2,2)/(120*math.pi*10))
    ax6.figure.canvas.draw() # update plot

def convert_E_to_H_U(ax4):
    """
    Scale/update H-field axis according to E-field axis.
    """
    y1, y2 = ax4.get_ylim()
    
    # scale the y-axis range of the H-field axis (ax2)
    ax5.set_ylim(y1/(120*math.pi), y2/(120*math.pi))
    ax5.figure.canvas.draw() # update plot

# set up data sets

# IEEE Std C95.1, 1999 Edition -- MPE for controlled environment

# E field -- Electric Field Strength (E) V/m
Ex_99 = [.003,0.1,3,30,100,300,3000,15000,300000]
Ey_99 = [614,614,614,61.4,61.4,61.4,194.2,194.2,194.2]

# H field -- Magnetic Field Strength (H) A/m
Hx_99 = [.003,0.1,100,300,3000,15000,300000]
Hy_99 = [163,163,.163,.163,.515,.515,.515]

# IEEE Std C95.1, 2005 Edition -- MPE for controlled environment

# E field -- Electric Field Strength (E) V/m
Ex_05 = [.003,0.1,1,30,100,300,3000,300000]
Ey_05 = [1842,1842,1842,61.4,61.4,61.4,194.2,194.2]

# H field -- Magnetic Field Strength (H) A/m
Hx_05 = [.003,0.1,100,300,3000,300000]
Hy_05 = [163,163,.163,.163,.515,.515]

# IEEE Std C95.1, 1999 Edition -- MPE for uncontrolled environment

# E field -- Electric Field Strength (E) V/m
Ex_99U = [.003,1.34,30,300,15000,300000]
Ey_99U = [614,614,27.5,27.5,194.16,194.16]

# H field -- Magnetic Field Strength (H) A/m
Hx_99U = [.003,0.1,30,100,300,15000,300000]
Hy_99U = [163,163,(16.3/30),.0729,.0729,.515,.515]

# IEEE Std C95.1, 2005 Edition -- MPE for uncontrolled environment

# E field -- Electric Field Strength (E) V/m
Ex_05U = [.003,1.34,30,400,2000,100000,300000]
Ey_05U = [614,614,27.5,27.5,61.4,61.4,194.16]

# H field -- Magnetic Field Strength (H) A/m
Hx_05U = [.003,0.1,30,100,400,2000,100000,300000]
Hy_05U = [163,163,(16.3/30),.0729,.0729,.163,.163,.515]

fig, ax1 = plt.subplots()
plt.xscale('log')
plt.grid(True, which="both")

fig.suptitle('Combined Graphic Representation of IEEE RF Safety Guideline in Controlled Environment-- 1999 vs 2005', fontsize=20)

# automatically update ylim of ax2 and ax3 when ylim of ax1 changes
ax1.callbacks.connect("ylim_changed", convert_E_to_S)
ax1.callbacks.connect("ylim_changed", convert_E_to_H)

ax2 = ax1.twinx()
ax3 = ax1.twinx()

# ax1.plot(Ex_99, Ey_99, 'b-', Ex_05, Ey_05, 'b--')
line1 = ax1.plot(Ex_99, Ey_99, 'b-', label='E-field 99')
line2 = ax1.plot(Ex_05, Ey_05, 'b--', label='E-field 05')
ax1.legend()
ax1.set_xlabel('Frequency (MHz)')
ax1.set_yscale('log')
ax1.set_xlim(.003, 1000000)
ax1.set_ylim(10, 100000)
ax1.set_ylabel('E field (V/m)', color='b')
for tl in ax1.get_yticklabels():
    tl.set_color('b')

# ax2.plot(Hx_99,Hy_99,'r-', Hx_05, Hy_05, 'r--')
line3 = ax2.plot(Hx_99, Hy_99, 'r-', label='H-field 99')
line4 = ax2.plot(Hx_05, Hy_05, 'r--', label='H-field 05')
ax2.legend(loc=4)
ax2.spines['right'].set_position(('axes', 0))
ax2.set_yscale('log')
ax2.set_ylabel('H field (A/m)', color='r')
for tl in ax2.get_yticklabels():
    tl.set_color('r')

ax3.set_yscale('log')
ax3.set_ylabel('Power Density (mW/cm^2)', color='g')
for tl in ax3.get_yticklabels():
    tl.set_color('g')

fig2, ax4 = plt.subplots()
plt.xscale('log')
plt.grid(True, which="both")

fig2.suptitle('Combined Graphic Representation of IEEE RF Safety Guideline in Uncontrolled Environment-- 1999 vs 2005', fontsize=20)

# automatically update ylim of ax5 and ax6 when ylim of ax4 changes
ax4.callbacks.connect("ylim_changed", convert_E_to_S_U)
ax4.callbacks.connect("ylim_changed", convert_E_to_H_U)

ax5 = ax4.twinx()
ax6 = ax4.twinx()

# ax4.plot(Ex_99U, Ey_99U, 'b-', Ex_05U, Ey_05U, 'b--')
line5 = ax4.plot(Ex_99U, Ey_99U, 'b-', label='E-field 99')
line6 = ax4.plot(Ex_05U, Ey_05U, 'b--', label='E-field 05')
ax4.legend()
ax4.set_xlabel('Frequency (MHz)')
ax4.set_yscale('log')
ax4.set_xlim(.003, 1000000)
ax4.set_ylim(10, 100000)
ax4.set_ylabel('E field (V/m)', color='b')
for tl in ax4.get_yticklabels():
    tl.set_color('b')

# ax5.plot(Hx_99U,Hy_99U,'r-', Hx_05U, Hy_05U, 'r--')
line7 = ax5.plot(Hx_99U, Hy_99U, 'r-', label='H-field 99')
line8 = ax5.plot(Hx_05U, Hy_05U, 'r--', label='H-field 05')
ax5.legend(loc=3)
ax5.spines['right'].set_position(('axes', 0))
ax5.set_yscale('log')
ax5.set_ylabel('H field (A/m)', color='r')
for tl in ax5.get_yticklabels():
    tl.set_color('r')

ax6.set_yscale('log')
ax6.set_ylabel('Power Density (mW/cm^2)', color='g')
for tl in ax6.get_yticklabels():
    tl.set_color('g')

plt.show()
```
